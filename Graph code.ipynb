```python
import networkx as nx
import matplotlib.pyplot as plt
import re

def parse_log_output(log_output):
    paths = []
    # Regex to find numeric sequences inside brackets
    path_regex = re.compile(r'path=\[(.*?)\]')
    for line in log_output.split('\n'):
        match = path_regex.search(line)
        if match:
            path_str = match.group(1)
            path = [int(node) for node in path_str.split(", ") if node.isdigit()]
            paths.append(path)
    return paths

# Your log output as a multiline string
log_output = """
AntMessage{source=2, content='resource', pheromoneLevel=1.0, ttl=2, hitCount=0, path=[2]}
AntMessage{source=2, content='resource', pheromoneLevel=1.0, ttl=1, hitCount=0, path=[2, 0]}
AntMessage{source=2, content='resource', pheromoneLevel=1.0, ttl=0, hitCount=0, path=[2, 0, 1]}
AntMessage{source=1, content='resource', pheromoneLevel=1.0, ttl=2, hitCount=0, path=[1]}
AntMessage{source=1, content='resource', pheromoneLevel=1.0, ttl=1, hitCount=0, path=[1, 2]}
AntMessage{source=1, content='resource', pheromoneLevel=1.0, ttl=1, hitCount=0, path=[1, 0]}
AntMessage{source=1, content='resource', pheromoneLevel=1.0, ttl=0, hitCount=0, path=[1, 2, 0]}
AntMessage{source=2, content='resource', pheromoneLevel=1.0, ttl=1, hitCount=0, path=[2, 1]}
AntMessage{source=1, content='resource', pheromoneLevel=1.0, ttl=0, hitCount=0, path=[1, 0, 2]}
AntMessage{source=2, content='resource', pheromoneLevel=1.0, ttl=0, hitCount=0, path=[2, 1, 0]}
"""

# Parse log output to get paths
paths = parse_log_output(log_output)

# Create a directed graph
G = nx.DiGraph()

# Add directed edges based on paths
for path in paths:
    for i in range(len(path) - 1):
        G.add_edge(path[i], path[i+1])

# Draw the directed network with arrows
nx.draw(G, with_labels=True, node_size=700, node_color='lightblue', font_size=12, font_weight='bold', arrows=True)
plt.show()
```
